```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/content/company_sales_data (2).csv')
df
```

```python
#1. Read Total profit of all months and show it using a line plot
x=df.iloc[:, 0]
y=df.iloc[:, -1]
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000])
```

```python
# 2. Get total profit of all months and show line plot with the following Style properties
#  Generated line plot must include following Style properties:
#  Line Style dotted and Line-color should be red
#  Show legend at the lower right location.
#  X label name = Month Number
#  Y label name = Sold units number
#  Add a circle marker.
#  Line marker color as read
#  Line width should be 3

plt.xlabel('Month Number')
plt.ylabel('Solid Units Number')
plt.plot(x,y, linestyle='dotted', color='r', marker='o', linewidth=3, mfc='red')

```

```python
# 3. Read all product sales data and show it  using a multiline plot
#     square marker with color=green
fccreame=df['facecream']
fcwash=df['facewash']
toothpst=df['toothpaste']
bsoap=df['bathingsoap']
shampoo=df['shampoo']
most=df['moisturizer']
plt.plot(x, fccreame,  marker='s', mfc='green', color='red')
plt.plot(x,  fcwash, marker='s', mfc='green', color='red')
plt.plot(x,  toothpst,  marker='s', mfc='green', color='red')
plt.plot(x, bsoap, marker='s', mfc='green', color='red')
plt.plot(x,  shampoo, marker='s', mfc='green', color='red')
plt.plot(x,  most, marker='s', mfc='green', color='red')



```

```python
# lst=[fccreame,fcwash,toothpst,bsoap,shampoo,most]
# for i in lst:
#   plt.plot(x, i, marker='s', mfc='green', color='red')
a=df.iloc[:, 1:-2]
for i in a:
  plt.plot(x, a[i])
```

```python
# 4.Read toothpaste sales data of each month and show it using a scatterplot
plt.scatter(x, toothpst)
```

```python
# 5.Read face cream and facewash product sales data and show it using the bar chart with grid lin
# plt.bar(x, fccreame, color='green')
# plt.bar(x, fcwash, color='blue')

# plt.grid(color='red')
plt.bar([a-0.10 for a in x],fccreame,width=0.25,align='edge', label='face cream')
plt.bar([a+0.10 for a in x],fcwash,width=0.25,align='edge', label='face wash')
# plt.xlabel('Month Number')
# plt.ylabel('sales data')
plt.xticks(x)
plt.grid(linewidth=1,linestyle='dashed')
plt.title('Total Production')
plt.legend(loc='upper left')
```

```python
# 6.Read the total profit of each month and show it using the histogram to see the most common
prange=[150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist( y, prange)
```

```python
# 7.Calculate total sale data for last year for each product and show it using a Pie chart
products=['face cream', 'face wash', 'tooth paste', 'bathsoap', 'shampooo', 'mostizure']
sum=[fccreame.sum(),fcwash.sum(),toothpst.sum(), bsoap.sum(), shampoo.sum(), most.sum()]
plt.pie(sum, labels=products, autopct='%1.1f%%')
plt.legend()
```
