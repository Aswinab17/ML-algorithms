```python
x=df.iloc[:,:8].values                    #2dimensional   #we need to form it as a array so we need to pass .values
x
```

```python
y=df.iloc[:,-1].values             #yalways 1 directional
y

```

# **after completing this we need to convert the data into training and testing data**

```python
#training and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train

```

```python
x_test
```

#normalization the next step is
all input need to be in the same range
2 techniques for normalization are
1.standard scalar
2.MinMax Scalar
its our choice to select which technique

1.standard scalar equation --->z=(x-u)/s


```python
#normalization
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)      #z=(x-u)/s
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#predict
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
#performance evaluation
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
score=accuracy_score(y_test,y_pred)
score
```
