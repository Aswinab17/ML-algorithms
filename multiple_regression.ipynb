```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/advertising.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#seaborn
import seaborn as sns
sns.regplot(x=df['TV'],y=y,color='red')
```

```python
sns.regplot(x=df['Radio'],y=y,color='green')
```

```python
sns.regplot(x=df['Newspaper'],y=y,color='blue')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[210,34,50]]))
```

```python
df1=pd.DataFrame({'Actual value':y_test,'pred':y_pred,'Difference':y_test-y_pred})
df1
```

```python
print("slope is")
list(zip(x,model.coef_))
```

```python
print("constant is ",model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error
print("mae is",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print("mape is",mean_absolute_percentage_error(y_test,y_pred))
print("r2 is",r2_score(y_test,y_pred))
```
