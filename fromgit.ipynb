```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
df['State'].value_counts()
```

```python
sns.regplot(x=df['R&D Spend'],y=y,color='red')
```

```python
sns.regplot(x=df['Administration'],y=y,color='green')
```

```python
sns.regplot(x=df['Marketing Spend'],y=y,color='yellow')
```

```python
#state is object so encoding
#onehotEncoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual value':y_test,'pred':y_pred,'Difference':y_test-y_pred})
df1

```

```python
print('slope is')
list(zip(x,model.coef_))
```

```python
print('Model constant is',model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error,r2_score
print('mean_absolute error is ',mean_absolute_error(y_test,y_pred))
```

```python
print(r2_score(y_test,y_pred))
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test,y_pred))
```
